## TRAINING AND TESTING

```python
data['date'][32].year
```

```python
train = data[data['date'].apply(lambda x: x.year)<=2016]
test = data[data['date'].apply(lambda x: x.year)>2016]
```

```python
x_train = train.drop('total',axis=1) 
y_train = train['total']
```

```python
x_test = test.drop('total',axis=1) 
y_test = test['total']
```

```python
x_train = x_train.drop('date',axis=1)
x_test = x_test.drop('date',axis=1)
```

## RIDGE REGRESSION

```python
from sklearn.linear_model import Ridge as rd
from sklearn.model_selection import GridSearchCV as gsv
```

```python
ridge=rd()
parameters={'alpha':[1e-15,1e-10,1e-8,1e-5,1e-3,1,5,20,30,35,40,45,50,55,60]}
ridge_regressor=gsv(ridge,parameters,scoring='neg_mean_squared_error',cv=5)
ridge_regressor.fit(x_train,y_train)
```

```python
ridge_regressor.best_params_
```

```python
prediction = ridge_regressor.predict(x_test) 
```

```python
sns.distplot(y_test-prediction)
```

```python
from sklearn import metrics

print('MAE',metrics.mean_absolute_error(y_test,prediction))
print('MSE',metrics.mean_squared_error(y_test,prediction))
print('RMSE',np.sqrt(metrics.mean_squared_error(y_test,prediction)))
```

## LASSO REGRESSION

```python
from sklearn.linear_model import Lasso as ls 
```

```python
lasso=ls()
parameters={'alpha':[1e-15,1e-10,1e-8,1e-5,1e-3,1,2,3,4,5,20,30,35,50]}
lasso_regressor=gsv(lasso,parameters,scoring='neg_mean_squared_error',cv=5)
lasso_regressor.fit(x_train,y_train)
```

```python
lasso_regressor.best_params_
```

```python
prediction2=lasso_regressor.predict(x_test)
```

```python
sns.distplot(y_test-prediction2)
```

```python
from sklearn import metrics

print('MAE',metrics.mean_absolute_error(y_test,prediction2))
print('MSE',metrics.mean_squared_error(y_test,prediction2))
print('RMSE',np.sqrt(metrics.mean_squared_error(y_test,prediction2)))
```
